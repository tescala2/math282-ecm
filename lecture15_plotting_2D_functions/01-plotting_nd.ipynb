```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

np.set_printoptions(precision=2)
```

### The idea for plotting a function f(x) was to do the following:
$$
x=\begin{bmatrix}
x_0 \\ x_1 \\ x_2 \\ x_3  
\end{bmatrix}
\qquad y=\begin{bmatrix}
f(x_0) \\ f(x_1) \\ f(x_2) \\ f(x_3)  
\end{bmatrix} \qquad 
$$
### where the vector y is computed in one shot using the vectorized function f:
$$
y = f(x)
$$

```python
x = np.arange(-1,1.25,0.5)
print(x)
```

```python
y = x**2
print(y)
```

```python
plt.plot(x,y,'o')
plt.show()
```

### To plot the function f(x,y) the idea is the same. We want to do
### $$
Z = f(X,Y)
$$
### in one single shot with a vectorized function. To do this we do as follow:

$$
X = \begin{bmatrix}
x_0 & x_1 & x_2 & x_3 \\
x_0 & x_1 & x_2 & x_3 \\
x_0 & x_1 & x_2 & x_3\\
x_0 & x_1 & x_2 & x_3 \\
x_0 & x_1 & x_2 & x_3
\end{bmatrix} \quad 
Y = \begin{bmatrix}
y_0 & y_0 & y_0 & y_0 \\
y_1 & y_1 & y_1 & y_1 \\
y_2 & y_2 & y_2 & y_2\\
y_3 & y_3 & y_3 & y_3 \\
y_4 & y_4 & y_4 & y_4
\end{bmatrix} 
\quad
Z = 
\begin{bmatrix}
f(x_0,y_0) & f(x_1,y_0) & f(x_2,y_0) & f(x_3,y_0)\\
f(x_0,y_1) & f(x_1,y_1) & f(x_2,y_1) & f(x_3,y_1)\\
f(x_0,y_2) & f(x_1,y_2) & f(x_2,y_2) & f(x_3,y_2)\\
f(x_0,y_3) & f(x_1,y_3) & f(x_2,y_3) & f(x_3,y_3) \\
f(x_0,y_4) & f(x_1,y_4) & f(x_2,y_4) & f(x_3,y_4)
\end{bmatrix} 
$$

### Example: Let's plot $f(x,y) = x^2 e^y$ on $-1\le x \le 1$ and $-2 \le y \le 2$

```python
x = np.arange(-1,1.5,0.5)
x
```

```python
y = np.arange(-2,2.5,0.5)
y
```

```python
def fun(x,y):
    xx = x**2
    yy = y**2
    return xx, yy
```

```python
a,b = fun(2,3)
print(a)
print(b)
```

### Obviously in practice you will write a vectorized function:

```python
def f(x,y):
    z = x**2 * np.exp(y)
    return z
```

```python
x = np.arange(-1,1.1,0.1)
y = np.arange(-2,2.1,0.1)
X,Y = np.meshgrid(x,y)
```

```python
Z = f(X,Y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color = 'black')
plt.show()
```

# Plotting a function and its tangent plane

### Consider the function:
$$
f(x,y) = 2x^2 -3xy + 8y^2 + 2x - 4y + 4
$$
### In class we have seen that the linear approximation of $f(x,y)$ around (2,-1) is:
$$
f(x,y) \approx 34 + 13(x-2) - 26(y+1)
$$
### or expanding the right handside:
$$
f(x,y) \approx  13x - 26y - 18
$$


```python
def f(x,y):
    z = 2*x**2 -3*x*y + 8*y**2 + 2*x - 4*y + 4
    return z
```

```python
def lin_app(x,y):
    z = 13*x-26*y - 18
    return z
```

### We are going to make the plot on the window $-50 \le x \le 50$ and $-20 \le y \le 20$.

```python
x = np.linspace(-50,50,20)
y = np.linspace(-20,20,20)
X,Y = np.meshgrid(x,y)
```

```python
Z = f(X,Y)
Z_lin = lin_app(X,Y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color = 'black')
ax.plot_wireframe(X, Y, Z_lin, color = 'red')
plt.show()
```
