```python
import numpy as np
import matplotlib.pyplot as plt
```

# Problem 1

### plot $f(x) = x \log(x)$ on $[0.01,1.5]$.

```python
def f(x):
    y = x * np.log(x)
    return y
```

```python
x = np.linspace(0.01,1.5,1000)
y = f(x)
plt.plot(x,y)
plt.show()
```

# Problem 2

### Plot the logistic curve $$ \frac{1}{1+e^{- \alpha x}}   $$
### for $\alpha =1$, $\alpha=2$ and $\alpha=10$ on the same plot on the interval $[-4,4]$.
### Include the legend.

```python
def logistic(x,alpha):
    # IMPUT: x is an ndarray, 
    #        alpha is a python float
    # OUTPUT: y is an ndarray
    y = 1/(1+np.exp(-alpha*x))
    return y
```

```python
x = np.arange(-4,4,0.01)
y1 = logistic(x,1)
y2 = logistic(x,2)
y10 = logistic(x,10)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y10)
plt.legend(['alpha=1','alpha=2','alpha=10'])
plt.show()
```
