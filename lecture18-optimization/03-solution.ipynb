```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy import linalg
```

# Exercise 1

### a) Plot the quadratic
### $$
f(x,y) = \frac{1}{2} (4x^2-2xy+8y^2)+5x - 9y
$$
### with contour and contourf. 

```python
def f(x,y):
    z = 0.5*(4*x**2-2*x*y+8*y**2) + 5*x - 9*y
    return z
    
```

```python
x = np.linspace(-5,5,400)
y = np.linspace(-5,5,400)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
plt.contour(X, Y, Z,50)
plt.colorbar()
plt.show()
```

```python
plt.contourf(X, Y, Z,10)
plt.colorbar()
plt.show()
```

### b) Solve the linear system
\begin{align*}
&\frac{\partial f}{\partial x}(x,y) = 0 \\
&\frac{\partial f}{\partial y}(x,y) = 0 
\end{align*}

```python
A = np.array([[4,-1],[-1,8]])
b = np.array([-5,9])
linalg.solve(A,b)
```

### b)  Expand these two functions and write them as
### $$
a x^2 + bxy + cy^2
$$
### then plot them on $-5\le x \le 5$ and $-5\le y \le 5$ with contour(). You will see that the quadratic that has both positive and negative eigenvalues gives you a saddle.

$$ f(x,y) = 5x^2 + 2 xy - 5 y^2  \qquad g(x,y) = -5x^2 -2 xy  - 5 y^2 $$

```python
def f(x,y):
    z = 5*x**2+2*x*y-5*y**2
    return z
```

```python
def g(x,y):
    z = -5*x**2-2*x*y-5*y**2
    return z
```

```python
x = np.linspace(-5,5,400)
y = np.linspace(-5,5,400)
X,Y = np.meshgrid(x,y)
Z1 = f(X,Y)
Z2 = g(X,Y)
```

```python
plt.contour(X, Y, Z1,50)
plt.colorbar()
plt.show()
```

```python
plt.contour(X, Y, Z2,50)
plt.colorbar()
plt.show()
```

### c) Plot $f(x,y)$ and $g(x,y)$ on the window $-5\le x \le 5$ and $-5\le y \le 5$ with plot_wireframe

```python
x = np.linspace(-5,5,20)
y = np.linspace(-5,5,20)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color = 'black')
plt.show()
```
